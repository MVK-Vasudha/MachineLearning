Import Libraries

```python


import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

Read Dataset

```python

df = pd.read_csv('Enjoy sports.csv')
```

Label Encoding

```python

df = df.drop(columns=['Day'])

# b. Use Label Encoding to encode categorical data
label_encoder = LabelEncoder()
encoded_df = df.apply(label_encoder.fit_transform)
```

Take Encoded Variables

```python
X = encoded_df.drop(columns=['Decision'])
y = encoded_df['Decision']

```

Make a Decision Tree

```python

clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X, y)
```

Plot the Tree

```python
plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=X.columns, class_names=['No', 'Yes'])
plt.show()
```
