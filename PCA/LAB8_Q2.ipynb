```python
from PIL import Image
import numpy as np
from sklearn.preprocessing import StandardScaler

```

```python

img = Image.open('images.jpeg').convert('L') 
img_array = np.array(img)  
img_flattened = img_array.reshape(-1, img_array.shape[1])  
scaler = StandardScaler()
img_scaled = scaler.fit_transform(img_flattened)  
print(img_scaled)

```

```python

cov_matrix_img = np.cov(img_scaled.T)
print(cov_matrix_img)

```

```python

eigenvalues_img, eigenvectors_img = np.linalg.eig(cov_matrix_img)
print(eigenvalues_img)
print(eigenvectors_img)
```

```python

sorted_idx_img = np.argsort(eigenvalues_img)[::-1]  
eigenvalues_sorted_img = eigenvalues_img[sorted_idx_img]
eigenvectors_sorted_img = eigenvectors_img[:, sorted_idx_img]
print(eigenvalues_sorted_img)
print(eigenvectors_sorted_img)
```

```python

k_img = 50
eigenvectors_k_img = eigenvectors_sorted_img[:, :k_img]
print(eigenvectors_k_img)
```

```python
img_transformed = np.dot(img_scaled, eigenvectors_k_img)

print("Transformed Image Data with 50 components:")
print(img_transformed)
```
